```python
import pandas as pd
import re

def convert_text_csv(chat):
    # Apna WhatsApp export file ka path yaha do
    file_path = chat
    
    # File read karna
    with open(file_path, "r", encoding="utf-8") as f:
        data1 = f.read()
    
    # Regex pattern for WhatsApp messages
    msg_pattern = r"^(\d{2}/\d{2}/\d{2}),\s*(\d{1,2}:\d{2}\s*(?:am|pm)) - (.*)"
    
    messages = []
    current_date = None
    current_time = None
    current_sender = None
    current_message = []
    
    for line in data1.split("\n"):
        match = re.match(msg_pattern, line, re.IGNORECASE)
        if match:
            # Save previous message if any
            if current_date:
                messages.append([current_date, current_time, current_sender, " ".join(current_message)])
            
            current_date, current_time, content = match.groups()
            if ": " in content:
                current_sender, msg = content.split(": ", 1)
            else:
                current_sender, msg = None, content
            current_message = [msg]
        else:
            # Continuation of previous message
            current_message.append(line)
    
    # Add last message
    if current_date:
        messages.append([current_date, current_time, current_sender, " ".join(current_message)])
    
    # Create DataFrame
    df = pd.DataFrame(messages, columns=["Date", "Time", "Sender", "Message"])
 
    
    # Agar CSV chahiye to save karo
    return df
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import string

# Download required data
nltk.download('stopwords')
nltk.download('punkt')

stem = PorterStemmer()

# Custom unwanted tokens
unwanted_words = {"omit", "mention", "<", ">", "media"}

# Example text
text = ' '.join(list(df['Message'])).lower()

# Tokenize
tokens = word_tokenize(text)

# Clean tokens
tokens = [
    stem.stem(word)
    for word in tokens
    if word not in stopwords.words('english')  # remove stopwords
    and word not in unwanted_words             # remove custom words
    and word not in string.punctuation         # remove punctuation
    and word.isalpha()                         # keep only alphabetic words
]

tokens = [
    stem.stem(word)
    for word in tokens
    if word not in stopwords.words('english')  # remove stopwords
    and word not in unwanted_words             # remove custom words
    and word not in string.punctuation         # remove punctuation
    and word.isalpha()                         # keep only alphabetic words
]

```

```python
tokens
```

```python
word_count={}

for i in tokens:
    word_count[i]=word_count.get(i,0)+1
```

```python
import pandas as pd

word_count = pd.Series(word_count).sort_values(ascending=False)[:10]
plt.barh(word_count.index, list(word_count.values), color='orange')
plt.title('Most Common Words')
plt.xlabel('Counts')
plt.ylabel('Word')
plt.show()
```

```python
num = ''.join(df['Message'].values)  # Join all text together

def is_emoji(s):
    return any(
        0x1F600 <= ord(char) <= 0x1F64F or  # Emoticons
        0x1F300 <= ord(char) <= 0x1F5FF or  # Symbols & pictographs
        0x1F680 <= ord(char) <= 0x1F6FF or  # Transport & map symbols
        0x2600 <= ord(char) <= 0x26FF or    # Misc symbols
        0x2700 <= ord(char) <= 0x27BF or    # Dingbats
        0x1F900 <= ord(char) <= 0x1F9FF or  # Supplemental symbols and pictographs
        0x1FA70 <= ord(char) <= 0x1FAFF or  # Symbols and pictographs extended-A
        0x200D == ord(char)                 # Zero width joiner
        for char in s
    )

emoji_count = {}
for char in num:
    if is_emoji(char):
        emoji_count[char] = emoji_count.get(char, 0) + 1


```

```python
emoji_count
    
```

```python
emoji_count=pd.DataFrame(emoji_count.items(), columns=['Emoji', 'Counts']).sort_values(by='Counts', ascending=False)

```
